## SET UP

```python
import pandas as pd
import os
import glob
```

## RUTAS

```python
ruta_proyecto = 'C:/Users/Oscar/OneDrive - FM4/Escritorio/EVOLVE/Data Science/Proyecto-Master-DataScience-Evolve-OscarFraile'
ruta_resultados = ruta_proyecto + '/05_Resultados/'
```

## CARGA Y MERGE

Lee automáticamente todos los CSV de `05_Resultados` que tengan la columna `periodo`.

```python
ficheros = glob.glob(ruta_resultados + 'airbnb_*.csv')
print(f'Ficheros encontrados: {len(ficheros)}')
for f in ficheros:
    print(' -', os.path.basename(f))
```

```python
dfs = []
for f in ficheros:
    df_temp = pd.read_csv(f)
    if 'periodo' not in df_temp.columns:
        print(f'AVISO: {os.path.basename(f)} no tiene columna periodo — se omite')
        continue
    dfs.append(df_temp)
    print(f'{os.path.basename(f)}: {df_temp.shape[0]} filas — periodo: {df_temp.periodo.unique()}')

df_all = pd.concat(dfs, ignore_index=True)
print(f'\nDataset combinado: {df_all.shape}')
df_all.periodo.value_counts()
```

## EXPORT

```python
ruta_salida = ruta_resultados + 'airbnb_all_periodos.csv'
df_all.to_csv(ruta_salida, index=False)
print(f'Exportado: {ruta_salida}')
print(f'Filas totales: {df_all.shape[0]}')
print(f'Columnas: {df_all.shape[1]}')
```
