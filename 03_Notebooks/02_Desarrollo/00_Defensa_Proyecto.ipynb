## Estructura recomendada de la defensa

1. **Problema y oportunidad:** qué abordáis, para quién y por qué importa.
2. **Producto desarrollado:** demo funcional, usuario, sistema y resultado.
3. **Valor y utilidad:** mejora concreta frente al proceso actual.
4. **Datos utilizados:** origen, formato, calidad, límites y tratamiento previo.
5. **Flujo funcional:** entrada, procesamiento, lógica aplicada, salida y feedback.
6. **Resultados obtenidos:** qué produce el producto y cómo se interpreta.
7. **Limitaciones y siguientes pasos:** qué no resuelve aún y cómo evolucionaría.
